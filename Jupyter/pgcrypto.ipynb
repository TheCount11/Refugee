```python
from pathlib import Path
import geopandas as gp

gdf = gp.read_file(Path.cwd().parents[0] / "Refugee" / "Jupyter" / "countries.geojson")
gdf.head()
```

```python
de = gdf[gdf['name']=='Germany']
de['post_guid'] = de['post_guid'].apply(str)
de.head()
```

```python
import hashlib
from base64 import b64encode, b64decode
import os

salt = os.urandom(32) 

def hashing(word):
        hexkey = hashlib.pbkdf2_hmac(
            'sha256', # The hash digest algorithm for HMAC
            word.encode('utf-8'), # Convert the password to bytes
            salt, # Provide the salt
            100000
            ).hex()

        b64 = b64encode(bytes.fromhex(hexkey)).decode()
        return b64[:-1] 
    
de['post_hll'] = de['post_guid'].apply(hashing)
de['date_hll'] = de['post_publish_date'].apply(hashing)
de.head()
```
