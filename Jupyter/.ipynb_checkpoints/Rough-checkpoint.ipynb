gdf =gp.GeoDataFrame(df_raw, geometry=gp.points_from_xy(df_raw.long, df_raw.lat))

SELECT ST_X(ST_TRANSFORM(post_latlng,3857)) AS LONG, 
	   ST_Y(ST_TRANSFORM(post_latlng,3857)) AS LAT,
		post_publish_date,
		hashtags,
		post_body,
		emoji
		FROM topical.post WHERE post_language = 'en' 
		AND  LOWER(post_body) NOT LIKE '%#moriarty%' ;

```python
df_sent.head() 
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

df_sent[['neg', 'neu', 'pos', 'compound']] = df_sent['post_body'].apply(analyzer.polarity_scores).apply(pd.Series)
df_sent.head()
    
```

```python
df_sent.nlargest(10,['pos'])
```

```python
df_sent.nlargest(10,['neg'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid", context='talk')


for var in ['pos', 'neg', 'neu', 'compound']:
    plt.figure(figsize=(12,4))
    sns.histplot(df_sent.query("pos > neg")[var], bins=30, kde=False, 
                 color='green', label='Positive')
    sns.histplot(df_sent.query("neg > pos")[var], bins=30, kde=False, 
                 color='red', label='Negative')
    plt.legend()
    plt.title(f'Histogram of {var} by true sentiment');
```

```python
test_sent1 =df_sent.copy(deep=True)
test_sent1.head()
```

```python



def clean_emoji(x):
    if x == '{}':
        return ''
    else:
        return x[1:-1]



test_sent1['emoji'] = test_sent1['emoji'].apply(clean_emoji)
test_sent1['hashtags'] = test_sent1['hashtags'].apply(lambda x:x[1:-1])
test_sent1.head()  
    
    
```

```python
test_sent1.to_csv(Path.cwd().parents[0] / "CSV" / "Raw"/ "tweetsentiment_clean.csv")
```
