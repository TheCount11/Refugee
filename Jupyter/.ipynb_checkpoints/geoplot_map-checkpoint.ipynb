```python
#imports

from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from collections import Counter
import geopandas as gp
import re,emojis
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import geoplot as gplt
import geoplot.crs as gcrs
import contextily as ctx
from shapely.geometry import Point, Polygon
import warnings
warnings.filterwarnings('ignore')

#read data
df = pd.read_csv(Path.cwd().parents[0] / "Refugee" / "CSV" / "Raw"/ "all_tweets.csv")
df.head()
```

```python
#clean data

def clean_emoji(x):
    if x == '{}':
        return ''
    else:
        return x[1:-1]

df['emoji'] = df['emoji'].apply(clean_emoji)
df['hashtags'] = df['hashtags'].apply(lambda x:x[1:-1])
df.head()  
```

```python
# remove hashtags which are not the top 5

tags = ['migranti','refugees', 'refugiados', 'flüchtlinge', 'vluchtelingen','réfugiés']    
    
def mask(x):
    foo = set(x.split(',')).intersection(tags)
    if len(foo)==0:
        return False
    else:
        return True

df = df.where(df['hashtags'].apply(mask))  
df.dropna(subset = ['hashtags','post_body'],inplace =True)
df.head()
```

```python
#convert to geodataframe

gdf = gp.GeoDataFrame(df, geometry=gp.points_from_xy(df.long, df.lat))
gdf.head()
```

```python
gdf.drop(columns = ['long','lat'],inplace=True)
```

```python
#get custom polygons for the countries

world = gp.read_file(
    gp.datasets.get_path('naturalearth_lowres'),
    crs="epsg:4326")

de = world[world['name'] == "Germany"]
uk = world[world['name'] == "United Kingdom"]
fr = world[world['name'] == "France"]
esp = world[world['name'] == "Spain"]
it = world[world['name'] == "Italy"]


fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(1, 5)
fig.suptitle(
    'Areas to test for common visitors in the hll benchmark dataset')
for ax in (ax1, ax2, ax3, ax4, ax5):
    ax.set_axis_off()
ax1.title.set_text('DE')
ax2.title.set_text('UK')
ax3.title.set_text('FR')
ax4.title.set_text('ESP')
ax5.title.set_text('IT')
de.plot(ax=ax1)
uk.plot(ax=ax2)
fr.plot(ax=ax3)
esp.plot(ax=ax4)
it.plot(ax=ax5)

    
```

```python
#remove unnecessary polygons from France

fr = fr.explode().iloc[1:].dissolve(by='name')
fr.plot()
```

```python
gdf.set_crs(epsg=4326,inplace=True)
gdf.crs
```

```python
eu_tweets = gp.sjoin(gdf,world,how="left",op='within')
```

```python
eu_tweets.dropna(subset = ["index_right"],inplace =True)
eu_tweets
```

```python
eu_tweets.total_bounds
```

```python
#creating a polygon for custom EU shapefile clipping

bbox = eu_tweets.total_bounds

p1 = Point(bbox[0], bbox[3])
p2 = Point(bbox[2], bbox[3])
p3 = Point(bbox[2], bbox[1])
p4 = Point(bbox[0], bbox[1])

np1 = (p1.coords.xy[0][0], p1.coords.xy[1][0])
np2 = (p2.coords.xy[0][0], p2.coords.xy[1][0])
np3 = (p3.coords.xy[0][0], p3.coords.xy[1][0])
np4 = (p4.coords.xy[0][0], p4.coords.xy[1][0])

bb_polygon = Polygon([np1, np2, np3, np4])

eu_polygon = gp.GeoDataFrame(gp.GeoSeries(bb_polygon), columns=['geometry'])
```

```python
eu_polygon.set_crs(epsg=4326,inplace = True)

eu = gp.clip(world,eu_polygon)
eu.plot()
```

```python
ctx.providers.Stamen.keys()
```

```python
extent = eu.total_bounds
ax = gplt.webmap(eu, projection=gcrs.WebMercator(),
                 provider = ctx.providers.Stamen.TonerLite,
                 figsize =(35,25))

gplt.kdeplot(eu_tweets, ax=ax,n_levels=30, cmap='turbo',kwargs = {"alpha":"0.5"})
#plt.savefig('Spatial_denstiy_EU.jpeg',quality =95,bbox_inches = 'tight')
```
